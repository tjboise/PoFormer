```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('white_blocks_distribution.csv')




plt.hist(data['Number of White Blocks'], bins=20, edgecolor = 'black')

```

```python
# Count the frequency of each number of white blocks
block_counts = data['Number of White Blocks'].value_counts()

# Sort the counts by number of white blocks
block_counts_sorted = block_counts.sort_index()

# Plotting the bar plot
# plt.figure(figsize=(10, 6))
block_counts_sorted.plot(kind='bar',edgecolor = 'black',width=0.9)
plt.xlabel('Number of Potholes in each image')
plt.ylabel('Count')
plt.title('Distribution of Number of Potholes')
plt.savefig('num_of_pothole.png',dpi=600,bbox_inches = 'tight')
```

```python
# Convert all numerical values to strings before grouping
data['Number of White Blocks Grouped'] = data['Number of White Blocks'].apply(lambda x: '>5' if x > 5 else str(x))


# Count the frequency of each group
grouped_block_counts = data['Number of White Blocks Grouped'].value_counts()
print(grouped_block_counts)
# Sort the counts by group
grouped_block_counts_sorted = grouped_block_counts.sort_index()

# Plotting the new bar plot
# plt.figure(figsize=(10, 6))
grouped_block_counts_sorted.plot(kind='bar', width=0.9,edgecolor = 'black',)
plt.xlabel('Number of Potholes in each image')
plt.ylabel('Count')
plt.title('Distribution of Number of Potholes')
plt.xticks(rotation=0)

plt.savefig('num_of_pothole.png',dpi=600,bbox_inches = 'tight')
```

```python
import cv2
import numpy as np

# Load the images
ground_truth = cv2.imread('paper/groundtruth.png', 0)
prediction = cv2.imread('paper/PoFormer.png', 0)
lraspp = cv2.imread('paper/LRASPP.png', 0)

# Threshold the images to get binary images
_, binary_ground_truth = cv2.threshold(ground_truth, 127, 1, cv2.THRESH_BINARY)
_, binary_prediction = cv2.threshold(prediction, 127, 1, cv2.THRESH_BINARY)
_, binary_lraspp = cv2.threshold(lraspp, 127, 1, cv2.THRESH_BINARY)

# Calculate the number of white pixels
white_pixels_ground_truth = np.sum(binary_ground_truth)
white_pixels_prediction = np.sum(binary_prediction)
white_pixels_lraspp = np.sum(binary_lraspp)

# Calculate the percentage difference
percentage_difference_poformer = ((white_pixels_prediction - white_pixels_ground_truth) / white_pixels_ground_truth) * 100
percentage_difference_lraspp = ((white_pixels_lraspp - white_pixels_ground_truth) / white_pixels_ground_truth) * 100

print(f"PoFormer is {percentage_difference_poformer}% different from the ground truth in terms of white area.")
print(f"LRASPP is {percentage_difference_lraspp}% different from the ground truth in terms of white area.")
```
